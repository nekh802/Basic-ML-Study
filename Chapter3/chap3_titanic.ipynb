```python
#방법3
#transformMixin, baseEstimator 이용
#Custom transformer만들고 파이프라인으로 연결

from sklearn.base import BaseEstimator, TransformerMixin

class AgeTransformer(BaseEstimator, TransformerMixin):
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        X['Age'].fillna(X['Age'].mean(), inplace=True)
        X['Age'] = X['Age'].astype('int')
        return X
```

```python
age_transform = AgeTransformer()
age_transform.fit_transform(titanic_train)
titanic_train.info()
```

```python
#first name만 표시하는 변환기 만들기
from sklearn.base import BaseEstimator, TransformerMixin
class NameTransformer(BaseEstimator, TransformerMixin):
    
    def __init__(self):
        self.name_ls = []
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None) :
        for name_idx in range(len(X['Name'])):
            self.name_ls.append(X['Name'][name_idx].split(',')[0])
        X['Name'] = self.name_ls
        return X
```

```python
name_transform = NameTransformer()
name_transform.fit_transform(titanic_train).tail()
#age 컬럼은 살아있는 것을 확인하자.
```

```python
#두 변환기를 파이프라인으로 연결
from sklearn.pipeline import Pipeline

full_pipeline = Pipeline([
    ('age_transform', AgeTransformer()),
    ('name_transform', NameTransformer())
])
```

```python
trans_train = full_pipeline.fit_transform(titanic_train)
trans_train.tail()
```

## Transformation Pipelines

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler()),
])

titanic_num_tr = num_pipeline.fit_transform(titanic_train_num)
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(titanic_train_num)
cat_attribs = ["Pclass"] ######

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])

titanic_prepared = full_pipeline.fit_transform(titanic_train)
```
